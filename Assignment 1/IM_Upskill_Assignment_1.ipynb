```
import pandas as pd
pd.plotting.register_matplotlib_converters()
from matplotlib import pyplot as plt
from sklearn import preprocessing
from itertools import chain
import numpy as np
import seaborn as sns
import math
%matplotlib inline
```

```
##Here, n is the input size and k is a given constant
n=list(range(1,7,1))
k=4
```

**Logarithmic Complexity**

```
logarithmic_complexity=[np.log(item) for item in n]
sns.lineplot(n,logarithmic_complexity)
plt.xlabel('Input Size')
plt.ylabel('Complexity- O(logn)')
plt.title('Fig 1: Graph of O(logn)')
```

**Linear Complexity**

```
linear_complexity=[item for item in n]
sns.lineplot(n,linear_complexity)
plt.xlabel('Input Size')
plt.ylabel('Complexity- O(n)')
plt.title('Fig 2: Graph of O(n)')
```

**Quadratic Complexity**

```
quadratic_complexity=[item**2 for item in n]
sns.lineplot(n,quadratic_complexity)
plt.xlabel('Input Size')
plt.ylabel('Complexity- O($\mathregular{n^2}$)')
plt.title('Fig 3: Graph of O($\mathregular{n^2}$)')
```

**Cubic Complexity**

```
cubic_complexity=[item**3 for item in n]
sns.lineplot(n,cubic_complexity)
plt.xlabel('Input Size')
plt.ylabel('Complexity- O($\mathregular{n^3}$)')
plt.title('Fig 4: Graph of O($\mathregular{n^3}$)')
```

**Polynomial Complexity**

```
polynomial_complexity=[item**k for item in n]
sns.lineplot(n,polynomial_complexity)
plt.xlabel('Input Size')
plt.ylabel('Complexity- O($\mathregular{n^k}$)')
plt.title('Fig 5: Graph of O($\mathregular{n^k}$)')
```

**Comparing O(n), O(logn), O($n^2$), O($n^3$) and O($n^k$)**<br>

Before plotting all five complexities together, I have normalized them because without normalisation it becomes hard to visualise as O(logn) is insignificant compared to cubic or polynomial complexity.

```

##Logarithmic
array_log=np.array(logarithmic_complexity)
normalized_square=preprocessing.normalize([array_log])
multi_dimensional_list=normalized_square.tolist()
flatten_log=list(chain.from_iterable(multi_dimensional_list))

##Linear
array_linear=np.array(linear_complexity)
normalized_square=preprocessing.normalize([array_linear])
multi_dimensional_list=normalized_square.tolist()
flatten_linear=list(chain.from_iterable(multi_dimensional_list))

##Squared
array_quadratic=np.array(quadratic_complexity)
normalized_square=preprocessing.normalize([array_quadratic])
multi_dimensional_list=normalized_square.tolist()
flatten_sq=list(chain.from_iterable(multi_dimensional_list))

##Cubic
array_cubic=np.array(cubic_complexity)
normalized_square=preprocessing.normalize([array_cubic])
multi_dimensional_list=normalized_square.tolist()
flatten_cube=list(chain.from_iterable(multi_dimensional_list))

##Polynomial
array_polynomial=np.array(polynomial_complexity)
normalized_square=preprocessing.normalize([array_polynomial])
multi_dimensional_list=normalized_square.tolist()
flatten_poly=list(chain.from_iterable(multi_dimensional_list))

plt.figure(figsize=(18,8))
plt.xlabel('Input Size')
plt.ylabel('Nomralized Time Complexity')
plt.title('Fig 6: Comparison of different Big O notation')


sns.lineplot(n,flatten_log,label='log')
sns.lineplot(n,flatten_linear,label='Linear')
sns.lineplot(n,flatten_sq,label='Quadratic')
sns.lineplot(n,flatten_cube,label='Cubic')
sns.lineplot(n,flatten_poly,label='Polynomial')
```

**Conclusion**<br>

Big O-notation tells us, how the computational complexity of algorithm increases as input size increases. From Figure 6, we can see that as the input size increases i.e. approaches n=7 in this case, increase in computational complexity of O(logn) is lowest whereas growth rate of 0(${n^k}$) is highest. If we arrange the growth rate of these 5 complexities in ascending order from left to right, the left being the best complexity and the right being the worst complexity we get, <br>

**O(logn) > O(n) > O(${n^2}$) > O(${n^3}$) > O(${n^k}$)** <br>

So, logarithmic complexity i.e. O(logn) is the best complexity as it has the lowest growth rate with respect to input size. Similarly, O(n) has lower growth rate compared to O(${n^2}$) making O(n) better than O(${n^2}$) and less efficient compared to O(logn).
